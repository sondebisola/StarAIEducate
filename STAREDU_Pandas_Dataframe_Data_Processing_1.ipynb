###**8. Data Cleaning:**  


##Handling missing values and outliers

```python
# Either statement will delete the rows with no value for Name
#df = df.dropna(subset=['Name'])
df = df.loc[df.Name.notnull()]

df = df.reset_index(drop=True)
df
```

#####Handling outliers

Notice one age was 140, which is unreasonably high.  It was likely a typo, but it could throw off results, so let's drop it from the data set.

```python
# Any of the lines below will keep rows where 'Age' is less than or equal to 100 or 'Age' is NaN
#df = df[(df['Age'] <= 100) | df['Age'].isna()]
#df = df[(df['Age'] <= 100) | df['Age'].isnull()]
#df = df.query('Age <= 100 or Age.isnull()')
df = df.drop(df[df['Age'] > 100].index)

df = df.reset_index(drop=True)
df
```

##### Using the mean (average) value

Some of the ages were missing, so let's fill the missing ages with the average value.

```python
# Fill missing ages with the mean
mean_age = df['Age'].mean()
df.loc[:, 'Age'] = df['Age'].fillna(mean_age).round(1)

df
```

#####Fixing Data Manually

Two genders were missing.  Let's fix these manually.

```python
# Update 'Gender' column for 'Chris' to 'M'
df.loc[df['Name'] == 'Chris', 'Gender'] = 'M'

# Update 'Gender' column for 'Sri' to 'F'
df.loc[df['Name'] == 'Sri', 'Gender'] = 'F'

#df
```

#####Using the middle value (median)

One of the salaries is missing.  Let us use the median (middle) value for the missing salary.

```python
# Fill missing data with middle value
median = df['Salary'].median()
df.loc[:,'Salary'].fillna(median, inplace=True)

df
```

##### Using the most common value (mode)

One of the countries is missing.  Let us use the mode (most frequent) value for the missing country.

```python
# Fill missing data with the most frequent value
most_frequent = df['Country'].mode()[0]
df['Country'].fillna(most_frequent, inplace=True)

df
```

####Removing duplicates:

We notice that there are duplicate entries for Alice, so we remove the duplicate entry.

```python
df = df.drop_duplicates()

df = df.reset_index(drop=True)
df
```

####Replace a specified value with another value.

A typo was found where the name "Charlie" was mistakenly entered as "Chalrie".  Let's fix that.

```python
df.loc[df['Name'] == 'Chalrie', 'Name'] = 'Charlie'

df
```

####We could have done all of the prior cleaning steps in one step through Lambda functions and Method Chaining.

```python
df2 = pd.DataFrame(data)
df2['Salary']= [50111,76083, 50430, 70000, None, 21800, 204560, 90750, 75030, 50111, 110808, 18500, 123005, 142500, 33303, 94500, 55004, 69038, 64500, 70340, 87320]

# Using value_counts for missing data count
missing_values = df2.apply(lambda col: col.isna().sum())
print(missing_values, '\n\n')

# Using method chaining for data cleaning
df2 = (
    df2.dropna(subset=['Name'])
      .query('Age <= 100 or Age.isna()')
      .assign(Age=lambda x: x['Age'].fillna(x['Age'].mean()).round(1))
      .assign(Gender=lambda x: x['Gender'].fillna(x['Name'].map({'Chris': 'M', 'Sri': 'F'})))
      .assign(Salary=lambda x: x['Salary'].fillna(x['Salary'].median()))
      .assign(Country=lambda x: x['Country'].fillna(x['Country'].mode()[0]))
      .assign(Name=lambda x: np.where(x['Name'] == 'Chalrie', 'Charlie', x['Name']))
      .drop_duplicates()
)

# Using groupby for basic statistics
age_stats = df2.groupby('Gender')['Age'].describe()
print(age_stats, '\n\n')

# Displaying top rows based on salary
top_salaries = df2.sort_values(by='Salary', ascending=False).head(10)
print(top_salaries, '\n\n')

df2
```
