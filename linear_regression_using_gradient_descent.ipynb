```python
import numpy as np #numpy is used for numerical computations, especially for handling arrays and matrices.
import pandas as pd #pandas is used for data manipulation and analysis, providing data structures like DataFrames.
```

```python
# Generate 100 random data points for the independent variable X and the dependent variable y.
np.random.seed(42)  # Set seed for reproducibility
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1) #A random error term is added to y to simulate real-world data.
print(X)
print(y)
```

```python
 """
    Performs gradient descent to find the optimal weights for linear regression.

    Args:
        X (np.ndarray): The feature matrix.
        y (np.ndarray): The target variable.
        learning_rate (float): The learning rate for gradient descent.
        iterations (int): The number of iterations to run.

    Returns:
        tuple: The optimal weights (theta) and the cost history.
    """

def gradient_descent(X, y, learning_rate=0.01, n_iterations=1000):
    global X_b

    m = len(y)
    theta = np.random.randn(2, 1)  # Initialize weights randomly  and iteratively updates them using the gradient descent algorithm
    X_b = np.c_[np.ones((m, 1)), X]  # Add bias term to X  to allow for a non-zero intercept.
    print(X_b)

    for iteration in range(n_iterations):
        gradients = 2/m * X_b.T.dot(X_b.dot(theta) - y)
        theta = theta - learning_rate * gradients

    return theta


#This line calls the gradient_descent function to
#train the model and obtain the optimized weights (including the bias)
theta = gradient_descent(X, y)
```

```python
X_new = np.array([[0], [2]]) #containing two new data points, 0 and 2,
print(X_new)
X_new_b = np.c_[np.ones((2, 1)), X_new] #adds a bias term (a column of ones) to X_new
#np.c_ function is used to concatenate the column of ones and X_new horizontally
print(X_new_b)
y_predict = X_new_b.dot(theta) #np.c_ function is used to concatenate the column of ones and X_new horizontally
print(y_predict)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X_new, y_predict, "r-")
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression with Gradient Descent")
plt.show()  # To see the output, run the code.
X
```

##determine performance of my model

with a regression problem, common metrics include Mean Squared Error (MSE), Root Mean Squared Error (RMSE), and R-squared.

1. Mean Squared Error (MSE):MSE measures the average squared difference between your model's predictions and the and the actual values. It gives you an idea of the overall error of your model.
2. Root Mean Squared Error (RMSE): RMSE is the square root of MSE and is often preferred because it's in the same units as the target variable, making it easier to interpret.
3. R-squared (Coefficient of Determination): R-squared represents the proportion of variance in the target variable that is explained by your model. It ranges from 0 to 1, with higher values indicating a better fit.


Lower MSE/RMSE: Indicates better model performance, as it means the predictions are closer to the actual values.
Higher R-squared: Indicates a better fit, meaning your model explains more of the variance in the target variable.

```python
import numpy as np

def mean_squared_error(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)

def root_mean_squared_error(y_true, y_pred):
    return np.sqrt(mean_squared_error(y_true, y_pred))

def r_squared(y_true, y_pred):
      mean_y_true = np.mean(y_true)
      ss_total = np.sum((y_true - mean_y_true)**2)
      ss_residual = np.sum((y_true - y_pred)**2)
      r2 = 1 - (ss_residual / ss_total)
      return r2

   # Assuming 'y' contains the actual values for the 100 data points
mse = mean_squared_error(y, y_pred_training)
rmse = root_mean_squared_error(y, y_pred_training)
r2 = r_squared(y, y_pred_training)

print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
print(f"R-squared: {r2}") # model explains approximately 78% of the variance in the target variable
```
